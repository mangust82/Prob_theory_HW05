```python
# Проведите тест гипотезы. 
# Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя 
# односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр 
# оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.
from scipy import stats

print('Гипотеза H0: M1 = M2 Гипотеза H1: M1 < M2, где M2=17.5мм')
print('Уровень статистической значимости alpha=0,05, по условию задачи. ')
print('Выбираем за статистический критерий Z-критерий поскольку известно среднекв отклонение')

Zn = (17.5-17)/(2/10)
print(f'Тогда наблюдаемый критерий Zn = (X-M)/(sdt/N^0.5)={Zn}')
Zt = stats.norm.ppf(0.95)
print(f'Теоретический критерий ищем на уровне 1-alpha, поскольку правосторонний тест Zt = {Zt}')
print('Zn лежит правее Zt. Гипотеза H0 не подтвердилась средний диаметр больше утверждаемого')
# P1 = stats.norm.pdf(182, loc=174, scale=8)
# P2 = stats.norm.cdf(182, loc=174, scale=8)


```

```python
# Задача 3. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
# Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
# Известно, что их веса распределены нормально.Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна
# 99%? (Провести двусторонний тест.)

import numpy as np
list = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
print('Гипотеза H0: M1 = M2 Гипотеза H1: M1 != M2, где M1 = 200 г')
print('Уровень статистической значимости alpha=0,01, по условию задачи. ')
print('Выбираем за статистический критерий критерий Стьюдента, поскольку неизвестно среднекв отклонение генеральной выборки')
alpha = 0.01
X = np.mean(list)
print(f'Вычислим средний вес пачки в выборке X = {X}')
std = np.std(list,ddof=1)
print(f'А также стандартное несмещенное отклонение по выборке std = {std}')
Tn = (198.5-200)/(std/pow(10, 0.5))
print(f'Тогда наблюдаемый критерий Tn = (X-M)/(sdt/N^0.5)={Tn}')
T1 = stats.t.ppf(1-alpha/2,df=2*(10-1))
T2 = stats.t.ppf(alpha/2,df=2*(10-1))
print(f'Рассчитаем коэффициенты Т1 и Т2 для alpha = 0.01. Т1 = {T1} Т2 = {T2}')
print('Наблюдаемый критерий Tn не входит в критическую область для alpha = 0.01. Гипотезу H0 не отвергаем средний вес пачки равен 200')
```

```python
# Задача 4. Есть ли статистически значимые различия в росте дочерей?
# Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165
# Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160, 163
mum = [172, 177, 158, 170, 178,175, 164, 160, 169, 165]
girl = [173, 175, 162, 174, 175, 168, 155, 170, 160, 163]
print('Гипотеза H0: Md = Mm Гипотеза H1: Md != Mm')
print('Выберем уровень статистической значимости alpha=0,05. ')
T, Pv =stats.ttest_ind(mum, girl)
print(f'Рассчитаем Pvalue с помощью функции в питоне. Pvalue = {Pv}')
print('Поскольку Pvalue больше alpha H0 не отвергаем. Следовательно статистически значимых различий в росте дочерей и матерей нет\n')

X1 = np.mean(mum)
X2 = np.mean(girl)
S1 = np.std(mum,ddof=1)
S2 = np.std(girl,ddof=1)
print(X1,X2)
print(S1,S2)
Tn = (X1-X2)/(pow(S1*S1/len(mum)+S2*S2/len(girl), 0.5))
print(f'Tn={Tn}')
```
